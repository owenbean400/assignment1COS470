```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

# Characters not needed for words
badCharacters = [".", "'", ":", "(", ")", "?"]
string = ""
depth = 30 ## Top 30 most frequent words

# Open file with Questions
with open("Questions.txt", "r+") as file:
    string = file.read()

string = string.lower()

# Remove bad characters
for character in badCharacters:
    string = string.replace(character, "")

words = string.split()

wordDictionary = {}

# Create dictionary of with frequency count
for word in words:
    if word in wordDictionary.keys():
        wordDictionary[word] += 1
    else:
        wordDictionary[word] = 1

# Sort dictionary by count descending
sortedWordDictionaryValues = list(dict(sorted(wordDictionary.items(), key=lambda item: item[1], reverse=True)[0:depth]).values())

maxValue = sortedWordDictionaryValues[0]

# Turn into percent function
def percentify(value, max):
    return round(value / max * 100)

# y axis values of Questions frequency
yAxis = [percentify(value, maxValue) for value in sortedWordDictionaryValues]

# x axis of the top 30 word frequency
xAxis = [str(number) for number in range(1, depth+1)]

# Ideal Zipf's law for y axis
ziffianValues = [100/i for i in range(1, depth+1)]

# Plot the line graph
plt.plot(list(range(0, depth)), yAxis, color="red", label="Questions")
plt.plot(list(range(0, depth)), ziffianValues, color="blue", label="Ziffian Values")
plt.legend()
plt.show()
```
